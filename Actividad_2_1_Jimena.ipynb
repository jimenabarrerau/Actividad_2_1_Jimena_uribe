```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Ventas_totales.csv')
df.info()
```

```python
df.head()
```

```python
valores_nulos=df.isnull().sum()
valores_nulos
```

```python
df2= df.copy()
df2
```

```python
#Imputación con mediana para columnas tarjetas_credito, tarjetas_debito, salon_ventas, subtotal_ventas_alimentos_bebidas
df2["tarjetas_credito"]=df2["tarjetas_credito"].fillna(round(df["tarjetas_credito"].median(),1))
df2["tarjetas_debito"]=df2["tarjetas_debito"].fillna(round(df["tarjetas_debito"].median(),1))
df2["salon_ventas"]=df2["salon_ventas"].fillna(round(df["salon_ventas"].median(),1))
df2["subtotal_ventas_alimentos_bebidas"]=df2["subtotal_ventas_alimentos_bebidas"].fillna(round(df["subtotal_ventas_alimentos_bebidas"].median(),1))
```

```python
valores_nulos = df2.isnull().sum()
valores_nulos
```

```python
#Imputación con media
df2["otros_medios"] = df2["otros_medios"].fillna(round(df["otros_medios"].mean(),1))
df2["bebidas"] = df2["bebidas"].fillna(round(df["bebidas"].mean(),1))
df2["almacen"] = df2["almacen"].fillna(round(df["almacen"].mean(),1))
df2["panaderia"] = df2["panaderia"].fillna(round(df["panaderia"].mean(),1))
df2["lacteos"] = df2["lacteos"].fillna(round(df["lacteos"].mean(),1))
```

```python
valores_nulos = df2.isnull().sum()
valores_nulos
```

```python
#Imputación de valor en específico
df2["otros"]=df2["otros"].fillna(0)
```

```python
valores_nulos = df2.isnull().sum()
valores_nulos
```

```python
#Imputación con Ffill
df2["carnes"]=df2["carnes"].fillna(method="ffill")
df2["verduleria_fruteria"]=df2["verduleria_fruteria"].fillna(method="ffill")
df2["alimentos_preparados_rotiseria"]=df2["alimentos_preparados_rotiseria"].fillna(method="ffill")
```

```python
#Imputación con Bfill
df2["indumentaria_calzado_textiles_hogar"]=df2["indumentaria_calzado_textiles_hogar"].fillna(method="ffill")
df2["electronicos_articulos_hogar"]=df2["electronicos_articulos_hogar"].fillna(method="ffill")
```

```python
valores_nulos = df2.isnull().sum()
valores_nulos
```

```python
df2.to_csv('Ventas_totales2')
```
